### Создания новой таблицы, где каждая строка включает заболевание, ген, PUBMEDID, и все остальные связанные характеристики

Мы фильтруем данные по ключевым словам для сердечно-сосудистых заболеваний.

Затем сохраняем все строки, которые содержат значения PUBMEDID, а также все связанные с ними характеристики.

Каждая строка новой таблицы будет содержать: заболевание, ген, PUBMEDID и все остальные данные, которые относятся к этим строкам.

```python
import pandas as pd

# Фильтрация данных по сердечно-сосудистым заболеваниям
cardiovascular_keywords = [
    'cardiovascular', 'heart', 'coronary', 'hypertension', 'stroke', 'myocardial', 
    'artery', 'vascular', 'atherosclerosis', 'arrhythmia', 'ischemic', 'angina', 
    'thrombosis', 'pulmonary embolism', 'atrial fibrillation', 'endocarditis', 
    'pericarditis', 'valve disease', 'venous', 'venous thromboembolism', 
    'peripheral artery disease', 'blood pressure', 'cholesterol'
]

# Отбор всех строк с сердечно-сосудистыми заболеваниями
cardiovascular_data = data[data['DISEASE/TRAIT'].str.contains('|'.join(cardiovascular_keywords), case=False, na=False)]

# Формирование новой таблицы с необходимыми колонками
# Колонки, которые будут в новой таблице
columns_needed = ['DISEASE/TRAIT', 'MAPPED_GENE', 'PUBMEDID', 'P-VALUE', 'INITIAL SAMPLE SIZE', 'REPLICATION SAMPLE SIZE', 'CHR_ID', 'CHR_POS', 'DATE ADDED TO CATALOG', 'FIRST AUTHOR', 'DATE', 'JOURNAL', 'LINK', 'STUDY', 'DISEASE/TRAIT', 'INITIAL SAMPLE SIZE', 'REPLICATION SAMPLE SIZE', 'REGION', 'CHR_ID', 'CHR_POS', 'REPORTED GENE(S)', 'MAPPED_GENE', 'UPSTREAM_GENE_ID', 'DOWNSTREAM_GENE_ID', 'SNP_GENE_IDS', 'UPSTREAM_GENE_DISTANCE', 'DOWNSTREAM_GENE_DISTANCE', 'STRONGEST SNP-RISK ALLELE', 'SNPS', 'MERGED', 'SNP_ID_CURRENT', 'CONTEXT', 'INTERGENIC', 'RISK ALLELE FREQUENCY', 'P-VALUE', 'PVALUE_MLOG', 'P-VALUE (TEXT)', 'OR or BETA', '95% CI (TEXT)', 'PLATFORM [SNPS PASSING QC]', 'CNV']

# Создание новой таблицы с указанными колонками
cardiovascular_summary = cardiovascular_data[columns_needed].copy()

# Сохранение новой таблицы
output_file = 'cardiovascular_summary.csv'
cardiovascular_summary.to_csv(output_file, index=False)

# Вывод первых строк таблицы для проверки
print(cardiovascular_summary.head())

# Сохранение таблицы в pandas DataFrame
```

```python
# Статистика 2: создание таблицы с коррелирующими значениями по каждому ССЗ
grouped_by_disease = cardiovascular_data.groupby('DISEASE/TRAIT').agg({
    'PUBMEDID': 'count',  # Количество записей для каждого заболевания
    'INITIAL SAMPLE SIZE': 'first',  # Примерная величина выборки
    'REPLICATION SAMPLE SIZE': 'first',  # Величина выборки для репликации
    'P-VALUE': 'mean'  # Среднее значение P-value для каждой болезни
}).reset_index()

# Показать результат пользователю
import ace_tools as tools; tools.display_dataframe_to_user(name="Cardiovascular Disease Statistics", dataframe=grouped_by_disease)

total_cardiovascular_records
```
